```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('phone.csv')
df.columns
```

```python
prices = df.groupby('phone_brand')['price'].mean().sort_values(ascending=False)
print(prices)
```

```python
prices.plot(kind='bar' , figsize=(10,5),title='Price of different phones')
plt.xlabel('phones')
plt.ylabel('Prices')
plt.show()
```

```python

brand_count = df['phone_brand'].value_counts()

print(brand_count)
```

```python

brand_price_range = df.groupby(['phone_brand', 'price_range'])['price'].mean().unstack()


brand_price_range.plot(kind='bar', figsize=(10, 5), title='Price Range by Phone Brand')
plt.xlabel('Phone Brand')
plt.ylabel('Average Price (USD)')
plt.show()
```

```python
#   
average_price = df['price'].mean()

#  
print(f"Average Price: {average_price:.2f} USD")
```

```python
#     
price_over_time = df.groupby('Year')['price_USD'].mean()

#  Line chart
price_over_time.plot(kind='line', figsize=(10, 5), title='Price Change Over Time')
plt.xlabel('Year')
plt.ylabel('Average Price (USD)')
plt.show()
```

```python
#      
brand_features = df.groupby('phone_brand')[['price_USD', 'BATTERY', 'ram', 'Display_Size']].mean()

#       
print(brand_features.sort_values(by='price_USD', ascending=False))
```
